```python
import pandas as pd

url = 'https://www.basketball-reference.com/leagues/NBA_2018_totals.html'
data = pd.read_html(url)[0]
```

```python
# filter out non-stat rows
data = data[data['Rk'] != 'Rk']
```

```python
cols = ['AST', 'TOV']
for c in cols:
    data[c] = data[c].astype('int')
    
sorted_by_asts = data.sort_values(by='AST', ascending=False)
```

```python
n = 100
top_asts = sorted_by_asts[:n]
top_asts.plot(y='TOV', x='AST', kind='scatter')
```

```python
top_asts.columns
```

```python
to_data = top_asts[['Player', 'Pos', 'G', 'AST', 'TOV']]
import numpy as np
to_data['Rk'] = np.arange(1, len(to_data) + 1)
to_data.set_index('Rk', drop=True, inplace=True)
to_data
```

```python
to_data.to_csv('../data/assists_tov.csv')
```

```python
pct_url = 'https://www.basketball-reference.com/leagues/NBA_2018_advanced.html'
pct_data = pd.read_html(pct_url)[0]
```

```python
pct_data = pct_data[pct_data['Rk'] != 'Rk']
pct_data.head()
```

```python
cols = ['AST%', 'TOV%', 'MP']
for c in cols:
    pct_data[c] = pct_data[c].astype('float')
    
sorted_by_asts = pct_data.sort_values(by='AST%', ascending=False)
# filter by median MP
sorted_by_asts = sorted_by_asts[sorted_by_asts['MP'] > 755]
```

```python
sorted_by_asts[:100].plot(y='TOV%', x='AST%', kind='scatter')
```
