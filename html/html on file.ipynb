```python
import requests
```

```python
def write_url(url, filename):
    
    filename = '../html/{0}.html'.format(filename)
    
    text = requests.get(url).text
    with open(filename, 'w') as f:
        f.write(text)
```

```python
write_url('https://www.basketball-reference.com/teams/MIL/stats_basic_ranks.html', 'bucks_basic_ranks')
```

```python
import pandas as pd
```

```python
def get_dataframe(filename, **options):
    
    filename = '../html/{0}.html'.format(filename)
    with open(filename, 'r') as f:
        return pd.read_html(f.read(), **options)
    
```
